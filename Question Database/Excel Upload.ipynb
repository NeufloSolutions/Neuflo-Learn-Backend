```python
def execute_sql(connection, sql):
    cursor = connection.cursor()
    try:
        cursor.execute(sql)
        connection.commit()
    except (Exception, psycopg2.DatabaseError) as error:
        print(error)
    finally:
        if cursor is not None:
            cursor.close()
```

```python
# SQL for creating tables
create_subjects_table_sql = """
CREATE TABLE IF NOT EXISTS Subjects (
    SubjectID SERIAL PRIMARY KEY,
    SubjectName TEXT NOT NULL
);
"""
```

```python
create_chapters_table_sql = """
CREATE TABLE IF NOT EXISTS Chapters (
    ChapterID SERIAL PRIMARY KEY,
    SubjectID INT NOT NULL,
    ChapterTitle TEXT NOT NULL,
    ChapterNumber INT NOT NULL,
    FOREIGN KEY (SubjectID) REFERENCES Subjects(SubjectID)
);
"""
```

```python
create_subtopics_table_sql = """
CREATE TABLE IF NOT EXISTS Subtopics (
    SubtopicID SERIAL PRIMARY KEY,
    ChapterID INT NOT NULL,
    SubtopicName TEXT NOT NULL,  -- Ensure this column name is correct
    FOREIGN KEY (ChapterID) REFERENCES Chapters(ChapterID)
);
"""
```

```python
# Execute SQL commands
execute_sql(conn, create_subjects_table_sql)
execute_sql(conn, create_chapters_table_sql)
execute_sql(conn, create_subtopics_table_sql)
```

```python
def insert_subject_data(conn, subject_name):
    sql = "INSERT INTO Subjects (SubjectName) VALUES (%s) RETURNING SubjectID;"
    cur = conn.cursor()
    cur.execute(sql, (subject_name,))
    subject_id = cur.fetchone()[0]
    conn.commit()
    cur.close()
    return subject_id
```

```python
def insert_chapter_data(conn, subject_id, chapter_number, chapter_title):
    sql = "INSERT INTO Chapters (SubjectID, ChapterNumber, ChapterTitle) VALUES (%s, %s, %s) RETURNING ChapterID;"
    cur = conn.cursor()
    cur.execute(sql, (subject_id, chapter_number, chapter_title))
    chapter_id = cur.fetchone()[0]
    conn.commit()
    cur.close()
    return chapter_id
```

```python
def insert_subtopic_data(conn, chapter_id, subtopic_name):
    sql = "INSERT INTO Subtopics (ChapterID, SubtopicName) VALUES (%s, %s);"  # Ensure these column names are correct
    cur = conn.cursor()
    cur.execute(sql, (chapter_id, subtopic_name))
    conn.commit()
    cur.close()
```

```python
def process_subject_data(conn, subject_name, subject_data):
    subject_id = insert_subject_data(conn, subject_name)
    for chapter_number, chapter_info in subject_data.items():
        chapter_title = chapter_info["title"]
        chapter_id = insert_chapter_data(conn, subject_id, chapter_number, chapter_title)
        for subtopic in chapter_info["Topics"]:
            if subtopic:  # Checks if the subtopic is not empty
                insert_subtopic_data(conn, chapter_id, subtopic)
```

```python

def execute_sql(connection, sql, data=None):
    cursor = connection.cursor()
    try:
        if data:
            cursor.execute(sql, data)
        else:
            cursor.execute(sql)
        connection.commit()
    except (Exception, psycopg2.DatabaseError) as error:
        logging.error(f"Error executing SQL: {error}")
        logging.error(f"SQL: {sql}")
        if data:
            logging.error(f"Data: {data}")
        connection.rollback()
    finally:
        if cursor is not None:
            cursor.close()

def preprocess_answer(answer):
    if isinstance(answer, str):
        answer = answer.strip().lower()
        # Splitting by comma to handle multiple answers and then sorting to maintain a standard order
        answer_parts = sorted(answer.replace(' ', '').split(','))
        # Joining back the sorted parts
        answer = ','.join(answer_parts)
    return answer

def insert_question(conn, row, current_file):
    try:
        row['Answer'] = preprocess_answer(row['Answer'])
        if len(row['Answer']) > 3:
            raise ValueError(f"Answer too long: {row['Answer']}")

        try:
            row['SubtopicID'] = int(row['SubtopicID'])
        except ValueError:
            raise ValueError(f"Invalid SubtopicID: {row['SubtopicID']}")

        # Convert the year integer to a date string in 'YYYY-01-01' format
        year_date = f"{row['Year']}-01-01"

        # Handle HasImage as a boolean
        has_image = row.get('HasImage', False)  # Assumes pandas reads it as boolean

        sql = """
        INSERT INTO Questions (ChapterID, SubtopicID, QuestionNo, Question, 
                               OptionA, OptionB, OptionC, OptionD, Year, Answer, 
                               Explanation, HasImage)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
        """
        values = (row['ChapterID'], row['SubtopicID'], row['QuestionNo'],
                  row['Question'], row['Option A'], row['Option B'],
                  row['Option C'], row['Option D'], year_date, row['Answer'],
                  row['Explanation'], has_image)
        execute_sql(conn, sql, values)
    except Exception as e:
        logging.error(f"Error inserting question no {row['QuestionNo']} from file {current_file}: {e}")
        logging.error(f"Row data: {row}")
        return False
    return True

def process_excel_files(conn, directory):
    for filename in sorted(os.listdir(directory)):
        if filename.endswith('.xlsx'):
            current_file = filename
            print(f"Processing file: {filename}")
            df = pd.read_excel(os.path.join(directory, filename))
            successful_rows = 0
            for _, row in df.iterrows():
                if insert_question(conn, row, current_file):
                    successful_rows += 1
            print(f"File upload completed: {filename}. Rows uploaded: {successful_rows}")
```

# **Excel File Check**

```python
import os
import openpyxl
import pandas as pd

def check_excel_files_for_embedded_images(folder_path):
    # Iterate through all files in the folder
    for file in os.listdir(folder_path):
        if file.startswith('~$'):  # Skip temporary files
            continue
        if file.endswith('.xlsx'):
            file_path = os.path.join(folder_path, file)
            print(f"Checking file: {file}")

            # Load the workbook
            try:
                workbook = openpyxl.load_workbook(file_path, data_only=True)
            except Exception as e:
                print(f"Error loading workbook: {e}")
                continue

            # Iterate through each sheet in the workbook
            for sheet_name in workbook.sheetnames:
                sheet = workbook[sheet_name]
                print(f"  Sheet: {sheet_name}")

                # Using pandas to get the shape of the data
                try:
                    df = pd.read_excel(file_path, sheet_name)
                    print(f"    Data shape: {df.shape}")
                except Exception as e:
                    print(f"Error reading sheet with pandas: {e}")
                    continue

                # Check for embedded images
                if sheet._images:
                    print("    Embedded images found")
                else:
                    print("    No embedded images found")
```

```python
# Example usage
folder_path = r"D:\Neuflo\Neuflo Learn Latest\neuflo dataset\neuflo dataset\excel\PHYS" # Replace with the path to your folder
check_excel_files_for_embedded_images(folder_path)
```

```python
import os
import pandas as pd

def validate_excel_columns(folder_path):
    # Define a function to validate the Year and Answer columns
    def is_valid_year(year):
        return str(year).isdigit() and len(str(year)) == 4

    def is_valid_answer(answer):
        return isinstance(answer, str) and len(answer) <=3

    # Iterate through all files in the folder
    for file in os.listdir(folder_path):
        if file.startswith('~$'):  # Skip temporary files
            continue
        if file.endswith('.xlsx') or file.endswith('.xls'):
            file_path = os.path.join(folder_path, file)
            print(f"Checking file: {file}")

            # Read the Excel file
            xls = pd.ExcelFile(file_path)

            # Iterate through each sheet in the Excel file
            for sheet_name in xls.sheet_names:
                df = pd.read_excel(xls, sheet_name)
                print(f"  Sheet: {sheet_name}")

                # Check if both Year and Answer columns exist
                if 'Year' in df.columns and 'Answer' in df.columns:
                    for index, row in df.iterrows():
                        if not is_valid_year(row['Year']) or not is_valid_answer(row['Answer']):
                            print(f"    Invalid data at row {index + 1}: Year = {row['Year']}, Answer = {row['Answer']}")
                else:
                    print("    'Year' and/or 'Answer' column not found")
```

```python
# Example usage
folder_path = r"D:\Neuflo\Neuflo Learn Latest\neuflo dataset\neuflo dataset\excel\PHYS"  # Replace with the path to your folder
validate_excel_columns(folder_path)
```

# **Checking the number of questions**

```python
import os
import pandas as pd

def count_rows_in_excel_files(directory):
    # List all files in the given directory
    files = os.listdir(directory)

    # Filter for Excel files (assuming .xlsx extension)
    excel_files = [f for f in files if f.endswith('.xlsx')]

    # Iterate through each Excel file and count the rows
    for file in excel_files:
        file_path = os.path.join(directory, file)
        try:
            # Read the Excel file
            df = pd.read_excel(file_path)

            # Get the number of rows
            num_rows = len(df)

            # Print file name and number of rows
            print(f"File: {file}, Number of Rows: {num_rows}")
        except Exception as e:
            print(f"Error reading {file}: {e}")
```
